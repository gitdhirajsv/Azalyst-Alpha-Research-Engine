## Section 8: Hard Rules (Non-Negotiable)

### Research Integrity Rules

1. **OOS IC is the only metric that matters.**  
   In-sample performance is meaningless. IC in-sample can be 0.30, OOS can be 0.00, and you have a broken model.

2. **Year 3 (or equivalent final test period) is never touched during training.**  
   - Not for validation
   - Not for hyperparameter tuning
   - Not for early stopping
   - Not for debugging
   - Never. Ever.
   - **Firm Standard:** Citadel (Purged K-Fold boundary integrity)

3. **Every feature must have its target computed at time T using only data available at T.**  
   No exceptions. If a feature is supposed to predict 1-hour returns, compute the return using close prices from the past-looking bar window only.
   - **Firm Standard:** Two Sigma (scientific method) + BlackRock (strict train/test)

4. **Purged K-Fold gap ≥ 48 bars minimum (4 hours at 5-min frequency).**  
   This is not configurable downward. If you reduce it to 24 bars, you have leakage.
   - **Firm Standard:** Citadel (AFML Ch.7) + RenTech (data breadth discipline)

5. **Cross-sectional label computation is mandatory.**  
   `alpha_label` and `future_ret_*` columns must be computed after pooling ALL symbols at time T, never per-symbol.
   - Example: at week 52, compute rank percentiles for all 443 symbols simultaneously, then assign quartile labels. Do NOT compute 443 separate labels.
   - **Firm Standard:** Two Sigma (cross-sectional regression)

6. **The 50-coin +8,111% result is under leakage investigation until explicitly cleared.**  
   Do not treat it as validated alpha. Do not build on it. Do not cite it in research decisions.

### Leakage Detection Rules

7. **If OOS IC >> IS IC, assume leakage first.**  
   Example: IS IC=0.05, OOS IC=0.30 → this is impossible without contamination. Investigate immediately.

8. **If `--force-invert` is required for positive returns, the signal is broken.**  
   Do not invert. Instead, investigate:
   - Is the label constructed correctly?
   - Are the predictions trading direction or magnitude?
   - Is there a sign flip in the PnL calculation?

9. **Feature cache staleness check before every backtest:**  
   - [ ] Every cache file has correct columns (72 features + required metadata)
   - [ ] Timestamps are aligned across all symbols
   - [ ] Coverage ≥ 80% of available data  
   - [ ] No NaN creep in critical columns
   - **Firm Standard:** D.E. Shaw (atomic decomposition)

    - IS IC vs OOS IC

    - Prediction distribution shift (KL divergence)
