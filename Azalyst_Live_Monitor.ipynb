# Azalyst Live Monitor

Open this notebook from the Azalyst repo root and run the cells below while the autonomous team is active.

- Cell 2 shows a one-time snapshot.
- Cell 3 refreshes the same display object every few seconds without blocking Jupyter output rendering.
- Stop the live loop with `Kernel -> Interrupt` in Jupyter.


```python
from IPython.display import HTML, display
from pathlib import Path
import sys
import time

ROOT = Path.cwd()
if not (ROOT / "monitor_dashboard.py").exists():
    raise FileNotFoundError("Open this notebook from the Azalyst repo folder.")

if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

import monitor_dashboard as md

monitor = md.PipelineMonitor()
REFRESH_SECONDS = 5
```

```python
display(HTML(md.build_dashboard_html(monitor.get_status())))
```

```python
display_handle = display(HTML(md.build_dashboard_html(monitor.get_status())), display_id=True)

while True:
    display_handle.update(HTML(md.build_dashboard_html(monitor.get_status())))
    time.sleep(REFRESH_SECONDS)
```
